# Matplotlib

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

## Plot

```python
# The syntax is plt.plot(x-axis,y-axis)
plt.plot([2,5,8,10],[1,3,2,6])
```

```python
x = np.linspace(-5, 5, 20)
y = np.sin(x)
plt.plot(x, y, 'bo-.')
```

```python
x = np.linspace(0, 10, 21)
z = x**2
plt.plot(x, z, 'r*:')
```

## Histogram

```python
from numpy.random import *
x = randn(1000)
plt.hist(x, bins=50);
```

## Scatter Plot

```python
x = randn(50)
y = rand(50)
plt.scatter(x, y, c='red', marker='*')

z = np.linspace(1, 3, 10)
w = 1/z
plt.scatter(z, w, c='green', alpha=0.5, marker='^');
```
